```python
import pandas as pd
import numpy as np
```

```python
# 서울 열린데이터 광장에서 원본 데이터 불러오기
df = pd.read_csv("./서울 부동산 실거래가.csv", encoding = 'euc-kr')
df
```

```python
df.info()
```

```python
# 새로운 컬럼 생성을 위한 앞의 4자리 숫자 추출 함수 정의
def extract_year(row):
    return str(row)[:4]

# '계약일' 컬럼에서 앞의 4자리 숫자 추출하여 새로운 컬럼 생성
df['연도'] = df['계약일'].apply(extract_year)
```

```python
# 연도 컬럼 문자형에서 숫자형으로 변경
df['연도'] = df['연도'].astype(int)
```

```python
df
```

```python
# 필터링할 연도 범위
start_year = 2016
end_year = 2022

# 연도 범위에 해당하는 데이터 필터링
df = df[(df['연도'] >= start_year) & (df['연도'] <= end_year)]
df
```

```python
# 사용할 연도 컬럼만 추출해서 새로운 데이터 프레임 생성
# 추출된 데이터프레임을 CSV 파일로 저장
df.to_csv('seoul__data.csv', index=False)
```

# 데이터 전처리

```python
df = pd.read_csv('./seoul__data.csv')
df
```

```python
df.columns
```

```python
# 타겟 변수 마지막으로 이동
target_column = '물건금액(만원)'  # 이동시킬 열 이름
df = df[[col for col in df.columns if col != target_column] + [target_column]]
```

```python
df = df.loc[df['취소일'].isnull()]
df
```

```python
# 필요없는 컬럼삭제
df = df.drop(columns=['자치구코드','법정동코드','지번구분','법정동명','지번구분명','본번','부번','토지면적(㎡)','권리구분',
                 '신고구분', '취소일','신고한 개업공인중개사 시군구명', '접수연도', '건물용도', '건물명'])
```

```python
# 추출된 데이터프레임을 CSV 파일로 저장
df.to_csv('Seoul_1.csv', index=False)
```
