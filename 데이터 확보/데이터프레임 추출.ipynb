```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./서울 부동산 실거래가.csv",encoding = 'euc-kr')
df
```

```python
df.info()
```

```python
# 필요없는 컬럼 삭제
df = df.drop(columns=['자치구코드','법정동코드','지번구분','지번구분명','본번','부번','토지면적(㎡)','권리구분','취소일',
                 '신고구분','신고한 개업공인중개사 시군구명'])
```

```python
# 새로운 컬럼 생성을 위한 앞의 4자리 숫자 추출 함수 정의
def extract_year(row):
    return str(row)[:4]

# '계약일' 컬럼에서 앞의 4자리 숫자 추출하여 새로운 컬럼 생성
df['연도'] = df['계약일'].apply(extract_year)
```

```python
df.drop('접수연도', axis = 1, inplace = True)
```

```python
df
```

```python
# 사용할 연도 컬럼만 추출해서 새로운 데이터 프레임 생성
seoul_df = df[df['연도'].isin(['2017', '2018', '2019', '2020', '2021', '2022', '2023'])]
seoul_df
# 추출된 데이터프레임을 CSV 파일로 저장
seoul_df.to_csv('seoul__data.csv', index=False)
```
