### EPU 지수 컬럼 합치기

```python
epu = pd.read_csv('EPU지수.csv')
epu
```

```python
epu['계약일'] = epu['Year'].astype(str) + '-' + epu['Month'].astype(str) 
```

```python
epu = epu[['계약일', 'South Korea']]
epu
```

```python
df['계약일'] = pd.to_datetime(df['계약일'], format='%Y-%m-%d')
epu['계약일'] = pd.to_datetime(epu['계약일'], format='%Y-%m-%d')
```

```python
df['월'] = df['계약일'].dt.to_period('M')
epu['월'] = epu['계약일'].dt.to_period('M')
```

```python
merged_df = pd.merge(df, epu, on='월', how='left')
```

```python
merged_df
```

```python
merged_df = merged_df.drop(columns = ['월','계약일_y'])
```

```python
merged_df
```

```python
merged_df = merged_df.rename(columns = {'계약일_x':'계약일'})
```

```python
merged_df.to_csv('all_df.csv', encoding='utf-8-sig', index=False)
```

```python
test_df = merged_df[merged_df['계약일'] >= '2020-01-01']
```

```python
train_df = merged_df[merged_df['계약일'] < '2020-01-01']
```

```python
test_df.reset_index(drop=True, inplace=True)
```

```python
train_df.reset_index(drop=True, inplace=True)
```

```python
test_df.to_csv('test_df.csv', encoding='utf-8-sig', index=False)
```

```python
train_df.to_csv('train_df.csv', encoding='utf-8-sig', index=False)
```
