```python
import pandas as pd
import numpy as np
```

```python
# test 데이터 불러오기
test_df = pd.read_csv('./DF_Preprocessing/test_df.csv')
test_df
```

```python
# 환율 데이터 불러오기
a = pd.read_csv('./DF_Preprocessing/USD_KRW 과거 데이터.csv')
a
```

- 환율 데이터 합치기

```python
# 필요한 컬럼만 추출
a = a[['날짜', '종가']]
```

```python
# 컬럼 이름 바꾸기
a.columns = ['계약일', '환율종가']
```

```python
# 시계열 변경
test_df['계약일'] = pd.to_datetime(test_df['계약일'])
a['계약일'] = pd.to_datetime(a['계약일'])
```

```python
test_df1 = pd.merge(test_df, a, on='계약일', how='left')
test_df1
```

```python
# 금값 데이터 불러오기
b = pd.read_csv('./DF_Preprocessing/금값 데이터.csv')
b
```

```python
# 컬럼이름 변경
b = b.rename(columns={'날짜' : '계약일', '종가' : '금값종가'})
```

```python
# 필요한 컬럼만 추출
b = b[['계약일', '금값종가']]
```

```python
# 시계열 변경
b['계약일'] = pd.to_datetime(b['계약일'])
```

```python
# 합치기
test_df2 = pd.merge(test_df1, b, on='계약일', how='left')
```

```python
test_df2
```

```python
# 선행 순환 지표 데이터 불러오기
c = pd.read_csv('./DF_Preprocessing/선행지수순환지표.csv', index_col=False)
c
```

```python
# 필요없는 데이터 삭제
c = c.iloc[2:86,:]
```

```python
# 컬럼명 바꾸기
c.columns = ['계약일', '동행지수순환변동치', '선행지수순환변동치']
c
```

```python
# 동행지수순환변동치 컬럼 삭제
c.drop('동행지수순환변동치', axis = 1, inplace=True)
```

```python
# 계약일 데이터에 있는 '\n' 삭제
c['계약일'] = c['계약일'].str.replace('\n', '')
```

```python
# 계약일 데이터에 있는 '월' 삭제
c['계약일'] = c['계약일'].str.replace('월', '')
```

```python
# 계약일 datetime으로 바꾸기
c['계약일'] = pd.to_datetime(c['계약일'], format = "%Y%m")
```

```python
# 월별 데이터 이므로 원본 데이터와 선행지수순환변동치 데이터의 월컬럼 만들기
test_df2['월'] = test_df2['계약일'].dt.to_period('M')
c['월'] = c['계약일'].dt.to_period('M')
```

```python
test_df2
```

```python
c.drop('계약일', axis = 1, inplace=True)
```

```python
# 데이터 프레임 합치기
test_df3 = pd.merge(test_df2, c, on='월', how='left')
test_df3
```

```python
# 데이터 합칠 때 사용한 컬럼이 월 컬럼과 OLS 결과 유의미하지 않은 값이 콜금리(1일물,평균) 컬럼 삭제
test_df3 = test_df3.drop(columns = ['월','콜금리(1일물,평균)'])
test_df3
```

- 추가적인 컬럼인 환율종가(일별), 금종가(일별), 선행지수순환변동치(월별) 컬럼 선택 후 OLS 결과시 콜금리삭제
