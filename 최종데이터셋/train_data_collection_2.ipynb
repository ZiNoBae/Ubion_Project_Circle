### 환율종가 데이터 합치기

```python
a = pd.read_csv('USD_KRW 과거 데이터.csv')
a
```

```python
# 필요한 컬럼만 추출
a = a[['날짜', '종가']]
a
```

```python
# 컬럼이름 변경
a = a.rename(columns={'날짜' : '계약일', '종가' : '환율종가'})
```

```python
# 시계열 변경
df['계약일'] = pd.to_datetime(df['계약일'])
a['계약일'] = pd.to_datetime(a['계약일'])
```

```python
# 합치기
merged_df1 = pd.merge(df, a, on='계약일', how='left')
```

```python
merged_df1
```

### 금값 데이터 합치기

```python
b = pd.read_csv('금값 데이터.csv')
b
```

### 선행지수 순환지표 컬럼 합치기

```python
c = pd.read_csv('선행지수순환지표.csv')
c
```

```python
c = c.iloc[2:86]
```

```python
c = c.reset_index(drop=True)
c
```

```python
# 컬럼 이름명 변경
c.columns = ['계약일', '동행지수순환변동치', '선행지수순환변동치'] 
```

```python
# 필요컬럼만 추출
c = c[['계약일', '선행지수순환변동치']]
c
```

```python
# 문자열 '월' 삭제
c['계약일'] = c['계약일'].str.replace('월', '')
```

```python
# 문자열 '\n' 삭제
c['계약일'] = c['계약일'].str.replace('\n', '')
```

```python
c
```

```python
# 시계열 변경
c['계약일'] = pd.to_datetime(c['계약일'], format='%Y%m')
```

```python
c
```

```python
# 월별 데이터 이므로 원본 데이터와 선행지수순환변동치 데이터의 월컬럼 만들기
merged_df2['월'] = merged_df2['계약일'].dt.to_period('M')
c['월'] = c['계약일'].dt.to_period('M')
```

```python
# 필요데잍 추출
c = c[['월', '선행지수순환변동치']]
```

```python
c
```

```python
# 데이터 합치기
merged_df3 = pd.merge(merged_df2, c, on='월', how='left')
```

```python
merged_df3
```

```python
# 데이터 합칠 때 사용한 컬럼이 월 컬럼과 OLS 결과 유의미하지 않은 값이 콜금리(1일물,평균) 컬럼 삭제
merged_df3 = merged_df3.drop(columns = ['월'])
```
